```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np

os.environ['PATH'] += r"C:\selenium drivers"
driver = webdriver.Chrome()
driver.get("https://www.cars24.com/buy-used-car?f=bodyType%3Ain%3Aluxury%20sedan&sort=bestmatch&serveWarrantyCount=true&gaId=131078919.1687810141&listingSource=TabFilter")
time.sleep(3)
height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if height == new_height:
        print("completed")
        break
    height = new_height
# names = driver.find_elements(By.CLASS_NAME, "_2lmIw")
# # for name in names:
#     print(name.text)
names = driver.find_elements(By.CLASS_NAME, "_2lmIw")
# for name in names:
#     print(name.text)
car_names = [name.text for name in names]


car_data = driver.find_elements(By.CLASS_NAME, "_13yb6")
car_info = [info.text.split('\n') for info in car_data]
car_dist = [sublist[0] for sublist in car_info]
car_owner = [sublist[1] for sublist in car_info]
car_fuel = [sublist[2] for sublist in car_info]
car_state = [sublist[3] if len(sublist) > 3 and sublist[3] != '' else np.nan for sublist in car_info]


ul_elements = driver.find_elements(By.CLASS_NAME, "_1hOnS")
car_geartype = []

for ul_element in ul_elements:
    elements = ul_element.find_elements(By.XPATH, ".//li[text()='Manual' or text()='Automatic']")
    car_geartype.extend([element.text for element in elements])



div_elements = driver.find_elements(By.CLASS_NAME, "_18ToE")

car_cost = []
for div_element in div_elements:
    span_element = div_element.find_element(By.TAG_NAME, "span")
    value = span_element.text
    car_cost.append(value)








        
```

```python
act_owner = []
for owner in car_owner:
    own = owner.split()[0]
    act_owner.append(own)
print(act_owner)
```

```python
len(car_cost)
```

```python
import numpy as np
dict2 = {'Name':np.array(car_names),'Distance':np.array(car_dist),'Owner':np.array(act_owner),'Fuel':np.array(car_fuel),'Location':np.array(car_state),'Drive':np.array(car_geartype),'Price':np.array(car_cost)}
df2 = pd.DataFrame.from_dict(dict2)
df2
```

```python
df2["Type"] = "Lux_sedan"
df2
```

```python
df2.to_csv("Lux_sedan.csv")
```

## just like this we have created csv files for 5 types of cars:- hatchbacks,SUV,sedans ,Luxury SUV and Luxury Sedans. Now we will combine all 5 of them.


```python
df3 = pd.read_csv("HatchBack.csv", index_col=[0])
df4 = pd.read_csv("sedan.csv", index_col=[0])
df5 = pd.read_csv("SUV.csv", index_col=[0])
df6 = pd.read_csv("Lux_SUV.csv", index_col=[0])
```

```python
df2
```

```python
car_24 = pd.concat([df2, df3, df4, df5, df6], axis=0)
car_24
```

```python
car_24 = car_24.sample(frac=1).reset_index(drop=True)
car_24
```

```python
# Split the 'Name' column into two separate columns
car_24[['Year', 'Car Name']] = car_24['Name'].str.split(' ', 1, expand=True)

# Remove any leading or trailing whitespace from the 'Car Name' column
car_24['Car Name'] = car_24['Car Name'].str.strip()

# Print the modified DataFrame
car_24
```

```python
car_24.drop('Name', axis=1, inplace=True)
car_24
```

```python
cars24 = car_24[['Car Name','Year','Distance','Owner','Fuel','Location','Drive','Type','Price']]
cars24
```

```python
cars24.to_csv("cars24_final.csv")
```

```python
cars24['Price'] = cars24['Price'].str.replace(',', '').str.replace('₹', '').astype(int)
cars24['Distance'] = cars24['Distance'].str.replace(',', '').str.replace('KM', '').astype(int)
cars24
```

```python
cars24['Owner'] = cars24['Owner'].str.replace('ST', '')
cars24
```

```python
cars24['Owner'] = cars24['Owner'].str.replace('ND', '')
cars24['Owner'] = cars24['Owner'].str.replace('RD', '')
cars24
```
